## **Insights**

### Insight 1: Lots of Support Calls Mean a Customer is Leaving

Customers who call customer service multiple times are highly likely to stop buying from us. A high number of service calls means the customer is frustrated by recurring problems, making them want to leave the platform.

### Insight 2: Older Members Stay Loyal Longer

Analysis of the Membership_Years feature shows that customers who have been with us for a long time rarely leave. Brand loyalty builds up over the years, meaning new sign-ups are the ones we need to worry about losing early on.

### Insight 3: If a Customer Hasn't Bought Anything Lately, They Are Probably Gone

The Days_Since_Last_Purchase feature shows that the longer a customer goes without buying something, the more likely they are to switch to a competitor. Long gaps between orders are a clear sign that a customer is losing interest.

### Insight 4: Leaving Items in the Cart is a Major Red Flag

Looking at the Cart_Abandonment_Rate reveals that customers who frequently add items to their shopping cart but walk away without paying are highly likely to drop out completely. This behavior shows they are losing the motivation to shop with us.

### Insight 5: Big Spenders Are Harder to Lose

Features like Lifetime_Value and Credit_Balance show that customers who have spent a lot of money with us in the past, or keep a cash balance in their account, are very reliable. They have invested too much in our store to leave easily.

### Insight 6: The Models Successfully Separate Good and Bad Customers

Both the Logistic Regression and Decision Tree models did a great job parsing the dataset. The final results prove that the models can accurately tell the difference between active shoppers and customers who are about to quit.

### Insight 7: The Decision Tree is the Clear Winner

The evaluation scores show that the Decision Tree model beat Logistic Regression across Accuracy, Recall, and F1-Score. The tree was much smarter at uncovering hidden patterns and catching customers right before they left, making it our best option.

## **Limitations**

* The model's accuracy depends on the quality and completeness of the dataset.
* Missing or incorrect data can reduce prediction accuracy.
* Since the model is trained on historical data, it may not always predict future customer behavior accurately.
* Customer preferences and market conditions can change over time, affecting the model's performance.
* The model should be updated regularly with new data to maintain good accuracy.

## **Possible Improvements**

* Identify customers who are likely to churn and take early action.
