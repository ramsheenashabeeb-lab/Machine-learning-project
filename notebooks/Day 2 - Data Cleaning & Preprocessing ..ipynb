# **Data Cleaning & Preprocessing**

The Ecommerce Customer Churn dataset contains missing values, duplicate records, inconsistent values, and categorical variables that must be cleaned before building machine learning models.

The objectives of this stage are:

* Handle missing values
* Remove duplicate records
* Fix inconsistent values
* Convert data types where required
* Encode categorical variables
* Prepare the dataset for machine learning

## **Load DataSet**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\user\Desktop\Machien learning Project\Data\ecommerce_customer_churn_dataset.csv")
df.head()
```

## **Checking Missing Values**

```python

df.isnull().sum().sort_values(ascending=False)
```

## **Handle Missing Values**

#### Fill Missing Numerical Values with Median

```python
num_cols = df.select_dtypes(include=['number']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

#### Fill Missing Categorical Values with Mode

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

The dataset contains missing values in both numerical and categorical columns. 

Missing numerical values were replaced using the median, and missing categorical values were replaced using the mode to preserve the dataset for further analysis.

## **Verifying Missing Values**

```python
df.isnull().sum().sort_values(ascending=False)
```

### Missing Value Treatment

The following actions were taken:

* Missing numerical values were filled using the median of each column.
* Missing categorical values were filled using the mode.
* After preprocessing, all missing values were successfully removed.
* These steps help maintain data quality and prepare the dataset for machine learning.

## **Check Duplicate Records**

```python
df.duplicated().sum()
```

## **Removing Duplicates**

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

## **Outlier Detection And Removing**

```python
outlier_columns = ['Churned', 'Membership_Years']

for col in outlier_columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    df = df[
        (df[col] >= lower_bound) &
        (df[col] <= upper_bound)
    ]
```

## **Fix Inconsistent Values**

```python
df[(df["Total_Purchases"] == 0) &
   (df["Average_Order_Value"] > 0)].shape
```

```python
df = df[~(
    (df["Total_Purchases"] == 0) &
    (df["Average_Order_Value"] > 0)
)]
```

## **Convert Data Types**

### **Check Data Types**

```python
df.dtypes 
```

## **Convert Numerical Columns to Integer**

```python
int_cols = [
    "Age",
    "Membership_Years",
    "Login_Frequency",
    "Wishlist_Items",
    "Total_Purchases",
    "Customer_Service_Calls",
    "Product_Reviews_Written",
    "Mobile_App_Usage",
    "Lifetime_Value",
    "Credit_Balance"
]

for col in int_cols:
    df[col] = df[col].astype(int)

df.dtypes
```

## **Categorical Columns**

```python
cat_cols = ["Gender", "Country", "City", "Signup_Quarter"]

for col in cat_cols:
    df[col] = df[col].astype("category")

df.dtypes
```

## **Check Cardinality Before Encoding**

```python
categorical_columns = ['Gender', 'Country', 'City', 'Signup_Quarter']

for col in categorical_columns:
    print(col, ":", df[col].nunique())
```

## **Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cols_to_encode = ['Country', 'City', 'Signup_Quarter']

for col in cols_to_encode:
    df[col] = le.fit_transform(df[col].astype(str))

df[cols_to_encode].head()
df
```

## **One-Hot Encoding**

```python
df= pd.get_dummies(
    df, 
    columns=['Gender'], 
    drop_first=True, 
    dtype=int
)
df.head()
```

```python
df.rename(columns={"Gender_Other":"Gender_Female"})
```
